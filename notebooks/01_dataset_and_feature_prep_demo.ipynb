```python
from pathlib import Path
import json
import subprocess

REPO_ROOT = Path.cwd().resolve()
print(REPO_ROOT)
```

```python
result = subprocess.run(
    ["python3", "scripts/demo_phase1_prep.py", "--config", "configs/data_prep_demo.json"],
    cwd=REPO_ROOT,
    capture_output=True,
    text=True,
    check=True,
)
print(result.stdout)
```
