# MASI Recommender Foundation Demo

This notebook exercises the first recommender-side implementation in the repository. It runs the synthetic recommender demo script, then inspects the summary artifact.

The underlying modules implement:
- SASRec-style sequential encoding,
- TIGER-style semantic-ID autoregressive training scaffolding,
- cross-modal MLM pretraining over fused `[TXT] ... [VIS] ...` token sequences.

```python
from pathlib import Path
import json
import subprocess

REPO_ROOT = Path.cwd().resolve()
PYTHON = REPO_ROOT / ".venv" / "bin" / "python"
print(REPO_ROOT)
print(PYTHON)
```

```python
result = subprocess.run(
    [str(PYTHON), "scripts/demo_recommender_foundation.py", "--config", "configs/recommender_amazon_csj_demo.json"],
    cwd=REPO_ROOT,
    env={**dict(), "PYTHONPATH": str(REPO_ROOT / "src")},
    capture_output=True,
    text=True,
    check=True,
)
print(result.stdout)
```

```python
summary_path = REPO_ROOT / "outputs" / "recommender_demo" / "recommender_demo_summary.json"
with summary_path.open("r", encoding="utf-8") as handle:
    summary = json.load(handle)
summary
```
